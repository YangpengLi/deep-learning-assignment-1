```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_svmlight_file

X, y = load_svmlight_file("scaled.txt")
X = pd.DataFrame(X.toarray())
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X.columns = features
y[y == -1] = 0

# Add the target variable as a new column in X
X['Outcome'] = y

diabetes_data = X[X['Outcome'] == 1]
no_diabetes_data = X[X['Outcome'] == 0]

feature1 = 'Age'
feature2 = 'BMI'

plt.figure(figsize=(10, 6))  
plt.scatter(diabetes_data[feature1], diabetes_data[feature2], c='red', label='Diabetes', alpha=0.5)
plt.scatter(no_diabetes_data[feature1], no_diabetes_data[feature2], c='blue', label='No Diabetes', alpha=0.5)

plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title(f'Scatter Plot of {feature1} vs. {feature2}')
plt.legend()
plt.grid(True)  
plt.savefig('linearly separable.png')  
plt.show()
```
